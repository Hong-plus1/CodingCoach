```python
from langchain_community.document_loaders import WebBaseLoader
from langchain.indexes import VectorstoreIndexCreator
from langchain_community.document_loaders import WebBaseLoader
from langchain.indexes import VectorstoreIndexCreator
from langchain.chains import RetrievalQA
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain.embeddings import OpenAIEmbeddings
import os
```

```python
# 加载环境变量
load_dotenv()

# 设置 USER_AGENT 环境变量
os.environ['USER_AGENT'] = 'MyApp/1.0'

# 创建 WebBaseLoader 实例，抓取网页内容
loader = WebBaseLoader("https://www.runoob.com/python/python-tutorial.html")

# 创建嵌入模型
embedding_model = OpenAIEmbeddings()

# 创建索引
index = VectorstoreIndexCreator(embedding=embedding_model).from_loaders([loader])
# index.query("python")
```

```python
# 创建 ChatOpenAI 模型
chat_model = ChatOpenAI(temperature=0)

# 创建 RAG 链
rag_chain = RetrievalQA(
    retriever=index.as_retriever(),
    llm=chat_model
)
# 用户提出的问题
user_question = "请解释一下Python中的列表解析是什么？"

# 使用 RAG 链回答用户的问题
response = rag_chain({"query": user_question})
print(response["result"])
```
